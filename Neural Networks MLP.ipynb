```python
x_train=np.random.rand(10000,10)
print(x_train)
```

```python
y=[0]*10000
for i in range(10000):
  for j in range(10):
    y[i] += (x_train[i][j]**(j+1))
    y=np.array(y)

print(y)
```

```python
model=tf.keras.models.Sequential()
```

```python
model.add(tf.keras.layers.Dense(10, activation=tf.nn.sigmoid)) #Introducing hidden layer 1.
model.add(tf.keras.layers.Dense(10, activation=tf.nn.sigmoid)) #Introducing hidden layer 2.
model.add(tf.keras.layers.Dense(1, activation=tf.keras.activations.linear)) #Final output layer.
```

```python
model.compile(optimizer='adam', 
              loss='mean_squared_error',metrics=['mean_squared_error'])
#Here Adam is adaptive learninng rate algorithm. 
#Loss is mean squared to avoid negative differences.
```

```python
model.fit(x_train, y, epochs=300)
```
